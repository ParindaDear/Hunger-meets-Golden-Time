# Promotion Strategy
**Objective:**    
I take the insights from the previous data analysis and the outcomes of predictive modeling, then turn them into practical strategies. To design promotions and improve operational processes in ways that can actually boost order volume and make deliveries more efficient

## Key Insights from Analysis
- **Golden Hours (17:00–23:00)** consistently generate the highest order volume across all food types.
- **Snack, Meal, and Drinks** have the highest overall order volume, while **Buffet** orders peak at specific times, especially around 19:00 and 22:00.
- **Off-peak hours (12:00–16:00)** have the lowest demand across all days.
- Predictive modeling confirms that **traffic conditions** and **order time** are the main factors affecting delivery time.
- **Festival days** and adverse **weather conditions** tend to slightly increase delivery time, but their impact is smaller compared to traffic and time of day.

## Context-Based Promotion Strategy
### Traffic Conditions 
- During high traffic periods, **avoid heavy promotions** to prevent too many orders at once.
- Focus on **rider allocation** to maintain service quality.

### Time-Based Promotion Strategy
##### Peak Hours (17:00–23:00)
- Focus promotions on **Snack, Meal, and Drinks**, since these food types are ordered the most during this period.
- Use **short and time-specific promotions**, such as limited deals around 19:00 and 22:00, when order volume is especially high
- Big discounts should be avoided during peak hours, as too many orders at once may affect delivery performance.
##### Off-Peak Hours (12:00–16:00)
- **Light discounts and Bundle deals** can be used to encourage more orders during this low-demand period.
- These promotions should mainly target **price-sensitive customers**, while keeping delivery workload manageable.

### Weather-Based Promotions
- On bad weather days such as stormy , offering **free delivery or lower delivery fees**. This helps reduce customer frustration when deliveries take longer than usual.
### Festival Promotions
- Use **time-limited promotions** during festival periods to capture higher demand.
